### [15 points] Problem 1: 
***

Consider the class C of concepts defined by triangles with **distinct** vertices of the form $(i, j)$ where $i$ and $j$ are integers in the interval $[0,99]$. A concept c labels points on the interior and boundary of the associated triangle as positive and points exterior to the triangle as negative.

**Note**: To make life easier, we'll allow degenerate triangles in $C$. That is, triangles where the vertices are collinear. The following image depicts an example of a nondegenerate and a degenerate triangle.

<img src="figs/triangles.png" width=400 height=50>  

**Part A**: Suppose we have an algorithm that produces a consistent $h$ from the hypothesis class $H = C$. Give a bound on the number of training examples sufficient to assure that for any target concept $c$ in $C$, our algorithm will, with probability $1-\delta$, output a hypothesis $h$ with generalization error at most $\epsilon$.

Notice that there a finite number of concepts in  CC . In the consistent case we can then use the bound

Since we have a finite number of concepts in C, with the consistent case we can use the formula:
<br>
$$
m \geq \frac{1}{\epsilon}\left(\ln\left| H \right| + \ln\frac{1}{\delta} \right)
$$

<br>
<br>
In a 100 X 100 grid, we have 10,000 points. Therefore, to form a triangle we need 3 points. 
So we have 10000C3 (10000 Choose 3), therefore $|H|$ is ${{10000}\choose{3}}$ which is $16616670000$
<br><br>
Now we know that $$\log|H| = \log{{10000}\choose{3}}$$ which is nothing but 
$$
\quad \Rightarrow \quad \log\left| H \right| \approx 25.83
$$
<br>
<br>
Then, our bound becomes 

$$
m \geq \frac{1}{\epsilon}\left(25.83 + \ln\frac{1}{\delta} \right) 
$$




**Part B**: Based on your bound in **Part A**, determine the minimum number of training examples necessary such that for any target concept $c$ in $C$, our algorithm will, with probability $0.95$, output a hypothesis $h$ with generalization error at most $0.15$.  

Substitute $\epsilon = 0.15$ and $\delta = 0.05$

$$
m \geq \frac{1}{0.15}\left(25.83 + \ln\frac{1}{0.05} \right) \approx 192.17
$$

This means we'll achieve our accuracy and confidence  with $m \geq 193$ training examples.

#### Proving the Lower Bound
- From the above diagrams we see that for 2 points, all dicotamies are satisfied. Therefore, the lower bound is VCdim(H) >= 2.
<br>
<br>

#### Proving the Upper Bound
##### First Hypothesis:
- Lets say we have 3 points, x1 x2 x3 at a distance r1, r2 and r3 from the center.
- Consider we have a circle of radius R. This is our hypothesis space.
- Lets assume that r1 <= r2 <= r3
- Consider the labels for points [x1, x2, x3] to be [y1, y2, y3] = [+1, -1, +1]
- Since x3 is positive i.e. y3 = +1 , this implies that R >= r3.
- But since r2 <= r3, this means that label of x2 should be positive but this is a contradiction as y2 = -1


##### Second Hypothesis:
- Since x2 is labelled -1, r2 < R. (Negative point lies inside the circle.)
- We assumed r1 <= r2. This implies that x1 lies inside the circle, so the label for x1 should be -1.
- This is a contradiction as y1 = +1




##### Result:
- From the above, we can conclude the  VCDim >= 2



### [20 points] Problem 3: Empirical Verification of PAC Bounds for Axis-Aligned Rectangles 
***

In the in-class notebook associated with PAC Learnability, we proved a PAC bound for the class of concepts $C$ comprised of axis-aligned rectangles living in $\mathbb{R}^2$ of the form $(a \leq x \leq b) \wedge (c \leq y \leq d)$ where $a, b, c, d$ are real numbers. Specifically, we proved that with probability $1-\delta$, any consistent learner could learn a hypothesis $h$ in $H = C$ with generalization error less than $\epsilon$ provided that the number of training examples satisfied 

$$
m > \frac{4}{\epsilon}\log\frac{4}{\delta}
$$

In this problem you will empirically verify this bound for the restricted concept class $C$ where the rectangles are defined by $(a \leq x \leq b) \wedge (c \leq y \leq d)$ where $a, b, c, d$ are real numbers satisfying $0 \leq a \leq b \leq 100$ and $0 \leq c \leq d \leq 100$.

- Write some code that computes approximate generalization errors for $100$ independent concepts $c$ and associated training sets of size $m$, and returns the worst-case generalization error at the confidence level $1-\delta$.  One way to do this in the case that say $\delta = 0.05$, is to report the $95^\textrm{th}$ percentile of the $100$ samples of the generalization error. We can then say that, in our simulation, $100(1-\delta)\%$ of our observed generalization errors were less than our computed value.
```python
import random
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math
%matplotlib inline

```

```python
class Coordinates:
    def __init__(self,x,y):
        self.x = x
        self.y = y

class Uniform_Model:
    def __init__(self,inputRange):
        self.inputRange = inputRange
    
    def generate_random_concept(self):
        coordinates = []
        coordinates.append(random.randint(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.randint(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.randint(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.randint(self.inputRange[0], self.inputRange[1]))
        coordinates.sort()
        
        self.concept = [Coordinates(coordinates[0],coordinates[2]), Coordinates(coordinates[1], coordinates[3])]
    
    def classify(self, test, boundary):
        if test.x >= boundary[0].x and test.x <= boundary[1].x and test.y >= boundary[0].y and test.y <= boundary[1].y:
            return 1
        else:
            return -1
    
    def generate_random_data(self, m_size):
        training_data = []        
        for i in range(m_size):
            x = random.randint(self.inputRange[0], self.inputRange[1])
            y = random.randint(self.inputRange[0], self.inputRange[1])
            training_data.append(Coordinates(x,y))            
        return training_data
    
    def get_hypothesis(self, training_data):
        small_x = 100000
        large_x = -1
        
        small_y = 100000
        large_y = -1
        for coor in training_data:
            if self.classify(coor, self.concept) == 1:
                if coor.x > large_x:
                    large_x = coor.x
                if coor.x < small_x:
                    small_x = coor.x
                if coor.y > large_y:
                    large_y = coor.y
                if coor.y < small_y:
                    small_y = coor.y 
        self.hypothesis = [Coordinates(small_x, small_y) , Coordinates(large_x, large_y)]
    
    def get_gen_error(self, concept_labels, hypothesis_labels):
        data_size = len(concept_labels)
        mis_match_count = 0
        
        for i in range(len(concept_labels)):
            if concept_labels[i] != hypothesis_labels[i]:
                mis_match_count +=1
        return mis_match_count/data_size

    def get_percentile(self, error_list, delta):
        return np.percentile(np.asarray(error_list), 100*(1-delta))
    
      
                                    
    def main_method(self, epochs, m_sizes, test_size, delta):
        delta_error = []
        for m_size in m_sizes:
            gen_error_list = []
            for epoch in range(epochs):            
                self.generate_random_concept()
                training_data = self.generate_random_data(m_size)
                self.get_hypothesis(training_data)
                test_data = self.generate_random_data(test_size)

                concept_labels =  []
                for t in test_data:
                    concept_labels.append(self.classify(t,self.concept))

                hypothesis_labels = []        
                for t in test_data:
                    hypothesis_labels.append(self.classify(t, self.hypothesis))

                gen_error = self.get_gen_error(concept_labels, hypothesis_labels)
                gen_error_list.append(gen_error)
                #print('For epoch ..',epoch,' ... for m =',m_size,' ... Gen error is: ', gen_error)
            
            #print(self.get_percentile(gen_error_list, delta))
            delta_error.append(self.get_percentile(gen_error_list, delta))
        return delta_error


model = Uniform_Model([0,100])
#m.main_method(1,[100], 1000)

model.main_method(100,[100,200], 1000, 0.05)

        
```

**Part B**: Use your code to estimate the generalization error with confidence parameter $\delta=0.05$ for training sets of size $m$ where $m = 250, 500, 1000, 1250,$ and $1500$ and the data are comprised of points $(x,y)$ where the $x$- and $y$-values are sampled from the continuous uniform distribution $\textrm{unif}(0,100)$. Make a **log-log** plot with $m$ on the horizontal axes and $\epsilon$ on the vertical axis.  Additionally, overlay the theoretical PAC bound on your graph and discuss your results. 

```python
def get_theoritical_gen_error(m_list, delta):
    thro_error = []
    for m in m_list:
        thro_error.append((4/m)*math.log(4/delta))
    return thro_error
    
```

```python
model = Uniform_Model([0,100])
#m.main_method(1,[100], 1000)

m_list = [250,500,1000,1250,1500]
delta = 0.05 
delta_error = model.main_method(100,m_list, 1000, delta)
plt.loglog(m_list,delta_error, label='practical_error')
#print(delta_error)

theo_error = get_theoritical_gen_error(m_list, delta)
plt.loglog(m_list,theo_error, label='theo_error')
plt.legend()
```

What is the difference between Normal distribution and Uniform Distribution?

- Uniform Distribution: Probability between the range is the same.
- Normal Distribution: Here it is the bell shaped curve. 
<br><br>
- What we mean is the following:
- Consider a normal distribution with mean = 50 and std deviation 25. So 1 std deviation is 25 to 75. Here 68% of the points will fall between the lines passing through (30,70).
- In uniform distribution, the points can fall randomly anywhere between the (0,100) and (100,0) graph
<br><br>
- So to solve this problem all that we need to do is substitute where ever random.randInt is used with random.normalvariate. The rest of the code since it is modular should take care of itself
<br><br>



```python
class Normal_Model:
    def __init__(self,inputRange):
        self.inputRange = inputRange
    
    def generate_random_concept(self):
        coordinates = []
        coordinates.append(random.normalvariate(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.normalvariate(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.normalvariate(self.inputRange[0], self.inputRange[1]))
        coordinates.append(random.normalvariate(self.inputRange[0], self.inputRange[1]))
        coordinates.sort()
        
        self.concept = [Coordinates(coordinates[0],coordinates[2]), Coordinates(coordinates[1], coordinates[3])]
    
    def classify(self, test, boundary):
        if test.x >= boundary[0].x and test.x <= boundary[1].x and test.y >= boundary[0].y and test.y <= boundary[1].y:
            return 1
        else:
            return -1
    
    def generate_random_data(self, m_size):
        training_data = []        
        for i in range(m_size):
            x = random.normalvariate(self.inputRange[0], self.inputRange[1])
            y = random.normalvariate(self.inputRange[0], self.inputRange[1])
            training_data.append(Coordinates(x,y))            
        return training_data
    
    def get_hypothesis(self, training_data):
        small_x = 100000
        large_x = -1
        
        small_y = 100000
        large_y = -1
        for coor in training_data:
            if self.classify(coor, self.concept) == 1:
                if coor.x > large_x:
                    large_x = coor.x
                if coor.x < small_x:
                    small_x = coor.x
                if coor.y > large_y:
                    large_y = coor.y
                if coor.y < small_y:
                    small_y = coor.y 
        self.hypothesis = [Coordinates(small_x, small_y) , Coordinates(large_x, large_y)]
    
    def get_gen_error(self, concept_labels, hypothesis_labels):
        data_size = len(concept_labels)
        mis_match_count = 0
        
        for i in range(len(concept_labels)):
            if concept_labels[i] != hypothesis_labels[i]:
                mis_match_count +=1
        return mis_match_count/data_size

    def get_percentile(self, error_list, delta):
        return np.percentile(np.asarray(error_list), 100*(1-delta))
    
      
                                    
    def main_method(self, epochs, m_sizes, test_size, delta):
        delta_error = []
        for m_size in m_sizes:
            gen_error_list = []
            for epoch in range(epochs):            
                self.generate_random_concept()
                training_data = self.generate_random_data(m_size)
                self.get_hypothesis(training_data)
                test_data = self.generate_random_data(test_size)

                concept_labels =  []
                for t in test_data:
                    concept_labels.append(self.classify(t,self.concept))

                hypothesis_labels = []        
                for t in test_data:
                    hypothesis_labels.append(self.classify(t, self.hypothesis))

                gen_error = self.get_gen_error(concept_labels, hypothesis_labels)
                gen_error_list.append(gen_error)
                #print('For epoch ..',epoch,' ... for m =',m_size,' ... Gen error is: ', gen_error)
            
            #print(self.get_percentile(gen_error_list, delta))
            delta_error.append(self.get_percentile(gen_error_list, delta))
        return delta_error


model = Normal_Model([0,100])
#m.main_method(1,[100], 1000)

model.main_method(100,[100,200], 1000, 0.05)
```
